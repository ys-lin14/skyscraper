```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image

from wordcloud import WordCloud
```

```python
idea_data = pd.read_csv('data/ideas.csv', parse_dates=['date'])
```

```python
num_rows, num_columns = idea_data.shape
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}\n')

columns = list(idea_data.columns)
print(f'Columns: {columns}\n')

dates = idea_data['date']
print(f'Dates range from {dates.min()} to {dates.max()}')
```

```python
text = ' '.join(idea_data['idea'])
sky_mask = np.array(Image.open('data/sky_mask.jpg'))
```

```python
word_cloud = WordCloud(
    background_color='white',
    contour_width=2,
    mask=sky_mask,
).generate(text)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
