```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image

from wordcloud import WordCloud
```

```python
wishlist_data = pd.read_csv('data/wishlist.csv')
```

```python
num_rows, num_columns = wishlist_data.shape

print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')

print(f'\nColumns: {list(wishlist_data.columns)}')
```

```python
wishlist_contains_na = wishlist_data.isna().any().any()
print(f'Wishlist contains missing values: {wishlist_contains_na}')

if (wishlist_contains_na): # drop missing values
    wishlist_data.dropna(inplace=True)
    print(f'Rows left after dropping NA: {wishlist_data.shape[0]}')
```

```python
text = ' '.join(wishlist_data['wish'])
sky_mask = np.array(Image.open('data/sky_mask.jpg'))
```

```python
word_cloud = WordCloud(
    background_color='white',
    contour_width=2,
    mask=sky_mask,
).generate(text)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
